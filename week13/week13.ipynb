```python
import matplotlib.pyplot as plt
import csv
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result1 = []
result2 = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '03':
            result1.append(float(row[-1]))
    if row[-2] != '':
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '03':
            result2.append(float(row[-2]))    

plt.plot(result1, color = 'red', label = '최고기온')
plt.plot(result2, color = 'blue', label = '최저기온')
plt.legend()
plt.show()
f.close()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            aug.append(float(row[-1]))
        elif row[0].split('-')[1] == '01':
            jan.append(float(row[-1]))    

        

plt.hist(aug, bins = 100, color = 'red', label = 'Aug')
plt.hist(jan, bins = 100, color = 'blue', label = 'Jan')

plt.legend()
plt.show()
f.close()
```
