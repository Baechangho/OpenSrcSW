```python
#실습6: copy
def cp(file1, file2):
    f = open(file1)
    lines = f.readlines()
    f.close()
    g = open(file2, 'w')
    for line in lines:
        g.write(line)
    g.close

cp('file1.txt', 'file2.txt')
```

```python
#실습7: wc
def wc(file1):
    f = open(file1)
    lines = f.readlines()
    f.close()
    i = 0
    c = 0
    for line in lines:
        i = i + 1
        sp = line.split()
        c = c+len(sp)
        
    print('num of lines:', i)
    print('num of Words:', c)
    
wc('file1.txt')
```

```python
#실습8: class_ex.py
class Calc:

    def sum(self, a, b):
        result = a + b
        print('{0:d} + {1:d} = {2:d} 입니다'.format(a,b, result))
    def sub(self, a, b):
        result = a - b
        print('{0:d} - {1:d} = {2:d} 입니다'.format(a,b, result))
    def multi(self, a, b):
        result = a * b
        print('{0:d} * {1:d} = {2:d} 입니다'.format(a,b, result))
    def divi(self, a, b):
        result = a / b
        print('{0:d} / {1:d} = {2:.0f} 입니다'.format(a,b, result))
```

```python
#실습9: 계산기 stack클래스
class Stack:
    
    def calculater(self, string):
        num_op = 0
        NUM = []
        OP = []
        string2 = string.split(' ')
        print (list(string2))
        for num in string2:
            num_op = num_op + 1
            #숫자저장
            if num_op % 2 == 1:
                NUM.append(int(num))
                
            #연산자저장
            elif num_op % 2 == 0:
                OP.append(num)
       
            
        NUM1 = []
        OP1 = []
        num_op = 0
        for num1 in string2:
            num_op = num_op + 1
            num_int = int(num_op / 2)
            #숫자저장
            if num_op % 2 == 1:
                print('number array:', NUM1)
                print('op array:', OP1)
                if num_op == 1:
                    NUM1.append(int(num1))
                else:
                    if(OP[0]=='*'or OP[0]=='/'):
                        if(num_op == 3):
                            print('')
                        else:
                            NUM1.append(int(num1)) 
                    else:
                        if(OP[int(num_op/2-1)]=='*'or OP[int(num_op/2-1)]=='/'):
                            print('')
                        else:
                            NUM1.append(int(num1))
            #연산자저장,계산
            elif num_op % 2 == 0:
                if (len(OP1) == 0):
                    if num1 == '+':
                        OP1.append(num1)
                        print('number array:', NUM1)
                        print('op array:', OP1)
                    elif num1 == '-':
                        OP1.append(num1)
                        print('number array:', NUM1)
                        print('op array:', OP1)
                    elif num1 == '*':
                        OP1.append(num1)
                        print('number array:', NUM1)
                        print('op array:', OP1)
                        NUM1[0] = NUM1[0] * NUM[num_int]
                        del OP1[0]
                    elif num1 == '/':
                        OP1.append(num1)
                        print('number array:', NUM1)
                        print('op array:', OP1)
                        NUM1[0] = NUM1[0] / NUM[num_int]
                        del OP1[0]
                elif (len(OP1) == 1):
                    if num1 == '+':
                        if OP1[0] == '+':
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[0] = NUM1[0] + NUM1[1]
                            del OP1[0]
                            del NUM1[1]
                        elif OP1[0] == '-':
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[0] = NUM1[0] - NUM1[1]
                            del OP1[0]
                            del NUM1[1]
                    elif num1 == '-':
                        if OP1[0] == '+':
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[0] = NUM1[0] + NUM1[1]
                            del OP1[0]
                            del NUM1[1]
                        elif OP1[0] == '-':
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[0] = NUM1[0] - NUM1[1]
                            del OP1[0]
                            del NUM1[1]
                    elif num1 == '*':
                        if OP1[0] == '-':
                            NUM1[1] = -NUM1[1]
                            OP1[0] = '+'
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[1] = NUM1[1] * NUM[num_int]
                            del OP1[1]
                        else:
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[1] = NUM1[1] * NUM[num_int]
                            del OP1[1]
                    elif num1 == '/':
                        if OP1[0] == '-':
                            NUM1[1] = -NUM1[1]
                            OP1[0] = '+'
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[1] = NUM1[1] / NUM[num_int]
                            del OP1[1]
                        else:
                            OP1.append(num1)
                            print('number array:', NUM1)
                            print('op array:', OP1)
                            NUM1[1] = NUM1[1] / NUM[num_int]
                            del OP1[1]
            
        if(len(OP1) == 1):
            print('number array:', NUM1)
            print('op array:', OP1)
            if OP1[0] == '+':
               NUM1[0] = NUM1[0] + NUM1[1]
            elif OP1[0] == '-':
               NUM1[0] = NUM1[0] - NUM1[1]
            del NUM1[1]
        print('answer:',NUM1[0])
        del NUM1[0]
```
